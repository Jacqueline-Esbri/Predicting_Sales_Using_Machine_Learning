```python
# Create year list
year_list = [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]

# Create blank dataframe
const_df = pd.DataFrame({
    "State":[],
    "Sex":[],
    "Race/Ethnicity":[],
    "Age Group":[],
    "First Year":[],
    "Last Year":[],
    "Rank":[],
    "Cause of Death":[],
    "Deaths":[]
})

# Loop through CSV for each year
for item in year_list:
    
    # Set file name
    fname = "resources/female_" + str(item) + ".csv"
    
    # Create dataframe for each year csv
    df = pd.read_csv(fname)
    
    # Loop through rows and drop "All Ages" rows
    for index,row in df.iterrows():
        if row[3] == "All Ages":
            df.drop(index, inplace=True)
    
    # Combine dataframes
    const_df = pd.concat([const_df, df])
    
# Drop State and Race/Ethnicity columns
const_df.drop(["State", "Race/Ethnicity"], axis=1, inplace=True)
    
# Write dataframe to new CSV file
const_df.to_csv('resources/female_2008_to_2018.csv', index=False)
```
